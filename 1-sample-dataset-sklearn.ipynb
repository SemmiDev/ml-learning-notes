```python
# load dataset
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
# karna dataset nya berupa dictionary, maka kita bisa mengaksesnya key nya dengan cara seperti ini
iris.keys()

# kita bisa melihat data nya dengan cara seperti ini
iris.data
# iris.feature_names
# iris.target_names
# iris.target
```

```python
# print desc
print(iris.DESCR)
```

```python
# explanatory variables (features)
X = iris.data
# X.shape
X # row column
```

```python
# response vairable (target)
y = iris.target
y
# y.shape
```

```python
# features_name
features_name = iris.feature_names
features_name

target_name = iris.target_names
target_name
```

```python
import matplotlib.pyplot as plt

X = X[:, :2]
x_min, x_max = X[:, 0].min() - 0.5, X[:,0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:,1].max() + 0.5

plt.scatter(X[:,0], X[:,1], c = y)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_max, y_min)
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=1)

print(f'X train: {X_train.shape}')
print(f'X test: {X_test.shape}')
print(f'y train: {y_train.shape}')
print(f'y test: {y_test.shape}')
```
